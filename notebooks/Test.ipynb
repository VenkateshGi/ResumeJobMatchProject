```python
from serpapi import GoogleSearch
import os
from dotenv import load_dotenv
load_dotenv()

params = {
    "q": "Senior AI Engineer LangChain jobs Hyderabad",
    "api_key": os.getenv("SERPAPI_KEY"),
    "num": 5
}

search = GoogleSearch(params)
results = search.get_dict()
print(results)
for r in results.get("organic_results", []):
    print(r['title'])
    print(r['link'])
    print(r.get('snippet', '')[:100])
    print("---")
```

```python
import google.generativeai as genai

# Configure with your API key
api_key = ""  # or use your API key directly
genai.configure(api_key=api_key)

# List available models
models = genai.list_models()
for model in models:
    print(f"Model: {model.name}")
    print(f"Display Name: {model.display_name}")
    print(f"Supported Generation Methods: {model.supported_generation_methods}")
    print("---")
```

```python
import google.generativeai as genai

# Configure with your API key
api_key = ""  # or use your API key directly
genai.configure(api_key=api_key)


# Filter and display embedding models
embedding_models = [m for m in genai.list_models() if 'embedding' in m.name.lower()]
for model in embedding_models:
    print(f"Model: {model.name}")
    print(f"Display Name: {model.display_name}")
    print(f"Supported Generation Methods: {model.supported_generation_methods}")
    print("---")
```
